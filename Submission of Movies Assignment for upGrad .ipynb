# Question10:Create a data visualization that shows each country and the number of top-rated movies produced in it. Find the country that produced the most top-rated movies.

```python
movies
```

```python
visualization=movies.groupby(['Country']).count()[["Title"]].sort_values(by="Title",ascending=False)
visualization
```

```python
visualization.plot(kind='bar')
plt.title("Number of top_rated movies by country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```
